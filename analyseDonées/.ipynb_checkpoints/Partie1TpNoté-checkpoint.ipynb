```python
# Séparation des données en ensembles d'apprentissage et de test
X_train, X_test, y_train, y_test = train_test_split(X_numpy, status_numpy, test_size=0.5, random_state=1)

# Affichage des dimensions pour confirmation
(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
def compare_models(X_train, y_train, X_test, y_test):
    # Initialisation des modèles
    cart = DecisionTreeClassifier(random_state=1)
    knn = KNeighborsClassifier(n_neighbors=5)
    mlp = MLPClassifier(hidden_layer_sizes=(40, 20), random_state=1)

    models = [cart, knn, mlp]
    model_names = ['CART', 'K-NN', 'MLP']
    results = {}

    for model, name in zip(models, model_names):
        # Entraînement du modèle
        model.fit(X_train, y_train)

        # Prédiction sur l'ensemble de test
        y_pred = model.predict(X_test)

        # Calcul des métriques
        confusion = confusion_matrix(y_test, y_pred)
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)

        # Enregistrement des résultats
        results[name] = {'Confusion Matrix': confusion,
                         'Accuracy': accuracy,
                         'Precision': precision,
                         'Recall': recall}

    return results

# Comparaison des modèles
model_comparison_results = compare_models(X_train, y_train, X_test, y_test)
model_comparison_results
```

```python
def find_best_model(results):
    best_accuracy_model = max(results, key=lambda x: results[x]['Accuracy'])
    best_precision_model = max(results, key=lambda x: results[x]['Precision'])
    best_recall_model = max(results, key=lambda x: results[x]['Recall'])

    return {
        "Best Accuracy Model": best_accuracy_model,
        "Best Precision Model": best_precision_model,
        "Best Recall Model": best_recall_model
    }

best_models = find_best_model(model_comparison_results)

best_models
```

```python
from sklearn.preprocessing import StandardScaler

# Initialiser le StandardScaler
normalisateur = StandardScaler()

# Ajuster le normalisateur sur les données d'entraînement et transformer les données d'entraînement et de test
X_train_normalise = normalisateur.fit_transform(X_train)
X_test_normalise = normalisateur.transform(X_test)

# Arbre CART sur les données normalisées
arbre_CART_normalise = DecisionTreeClassifier(random_state=1)
arbre_CART_normalise.fit(X_train_normalise, y_train)
score_arbre_CART_normalise = arbre_CART_normalise.score(X_test_normalise, y_test)

# k-Plus Proches Voisins sur les données normalisées
kppv_normalise = KNeighborsClassifier(n_neighbors=5)
kppv_normalise.fit(X_train_normalise, y_train)
score_kppv_normalise = kppv_normalise.score(X_test_normalise, y_test)

# Perceptron multicouche sur les données normalisées
perceptron_normalise = MLPClassifier(hidden_layer_sizes=(40, 20), random_state=1)
perceptron_normalise.fit(X_train_normalise, y_train)
score_perceptron_normalise = perceptron_normalise.score(X_test_normalise, y_test)

# Afficher les scores
print("Score de l'arbre CART normalisé :", score_arbre_CART_normalise)
print("Score du k-Plus Proches Voisins normalisé :", score_kppv_normalise)
print("Score du perceptron multicouche normalisé :", score_perceptron_normalise)
```

```python
from sklearn.decomposition import PCA

#l'ACP sur les données normalisées
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_normalise)
X_test_pca = pca.transform(X_test_normalise)

# Concaténer les dimensions de l'ACP avec les données normalisées
X_train_augmente = np.concatenate((X_train_normalise, X_train_pca), axis=1)
X_test_augmente = np.concatenate((X_test_normalise, X_test_pca), axis=1)

# Réappliquer les modèles sur les nouvelles données

# Arbre CART
arbre_CART_augmente = DecisionTreeClassifier(random_state=1)
arbre_CART_augmente.fit(X_train_augmente, y_train)
score_arbre_CART_augmente = arbre_CART_augmente.score(X_test_augmente, y_test)

# k-Plus Proches Voisins
kppv_augmente = KNeighborsClassifier(n_neighbors=5)
kppv_augmente.fit(X_train_augmente, y_train)
score_kppv_augmente = kppv_augmente.score(X_test_augmente, y_test)

# Perceptron multicouche
perceptron_augmente = MLPClassifier(hidden_layer_sizes=(40, 20), random_state=1)
perceptron_augmente.fit(X_train_augmente, y_train)
score_perceptron_augmente = perceptron_augmente.score(X_test_augmente, y_test)

(score_arbre_CART_augmente, score_kppv_augmente, score_perceptron_augmente)



```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, accuracy_score, precision_score


def custom_score(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='macro')  # Ensure to set average type based on your label distribution
    return (accuracy + precision) / 2

custom_scorer = make_scorer(custom_score)


param_grid = {
    'hidden_layer_sizes': [(100,)],  # Keeping only one architecture
    'solver': ['adam'],             # Only one solver
    # Removed 'activation' and 'learning_rate_init' to use default values
}


grid_search = GridSearchCV(
    MLPClassifier(max_iter=500, early_stopping=True, validation_fraction=0.1),  # Reduced max_iter
    param_grid, 
    scoring=custom_scorer, 
    cv=3,  # Kept the reduced number of cross-validation folds
    n_jobs=-1  # Use all available cores for parallel processing
)


grid_search.fit(X_train_normalise, y_train)

# Meilleurs paramètres et score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Meilleurs paramètres:", best_params)
print("Meilleur score:", best_score)
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
import numpy as np
import time
clfs = {
    'NaiveBayes': GaussianNB(),
    'CART': DecisionTreeClassifier(random_state=42),
    'ID3': DecisionTreeClassifier(criterion='entropy', random_state=42),
    'DecisionStump': DecisionTreeClassifier(max_depth=1, random_state=42),
    'MLP': MLPClassifier(hidden_layer_sizes=(20, 10), random_state=42),
    'RF': RandomForestClassifier(n_estimators=200, random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'AdaBoost': AdaBoostClassifier(n_estimators=200, random_state=42)
}


```

```python
# Modification de la fonction run_classifiers pour inclure le temps d'exécution et l'AUC

def run_classifiers_with_time_auc(clfs, X, Y):
    results = {}
    kf = KFold(n_splits=10, shuffle=True, random_state=0)
    for i in clfs:
        start_time = time.time()
        clf = clfs[i]

        # Calcul de la précision et de l'AUC par validation croisée
        cv_acc = cross_val_score(clf, X, Y, cv=kf, scoring='accuracy')
        cv_auc = cross_val_score(clf, X, Y, cv=kf, scoring='roc_auc')

        # Temps d'exécution
        execution_time = time.time() - start_time

        # Stockage des résultats
        results[i] = {
            'Accuracy': f"{np.mean(cv_acc):.3f} +/- {np.std(cv_acc):.3f}",
            'AUC': f"{np.mean(cv_auc):.3f} +/- {np.std(cv_auc):.3f}",
            'Execution Time': f"{execution_time:.2f} seconds"
        }

    return results

# Exécution de la fonction sur les ensembles de données non normalisés et normalisés
results_non_normalized = run_classifiers_with_time_auc(clfs, X_train, y_train)
results_normalized = run_classifiers_with_time_auc(clfs, X_train_normalise, y_train)

results_non_normalized, results_normalized



```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X[:, self.columns]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Assuming sorted_idx and optimal_num_variables are defined
optimal_num_variables = 13  

# Création du pipeline avec Naive Bayes
pipeline_nb = Pipeline([
    ('scaler', StandardScaler()),
    ('feature_selection', ColumnSelector(columns=sorted_idx[:optimal_num_variables + 1])),
    ('classifier', GaussianNB())
])

# Entraînement du pipeline avec Naive Bayes
pipeline_nb.fit(X_train, y_train)
```
